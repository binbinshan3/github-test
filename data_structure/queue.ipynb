```python
class Queue:
    def __init__(self):
        self.items=[]
    def isEmpyt(self):
        return self.items==[]
    def enqueue(self,item):
        return self.items.insert(0,item)
    def dequeue(self):
        return self.items.pop()
    def size(self):
        return len(self.items) 
```

```python
# 打印机建模问题 这个难
# 在一个小时内大概有十名学生，每名学生发起2次打印，每次1-20页
# 可以设定打印机速度 对该问题进行模拟
import random
class Printer():
    def __init__(self,ppm):
        self.speed=ppm
        self.currentTask=None
        self.timeRemaining=0
    def tick(self):
        if self.currentTask!=None:
            self.timeRemaining-=1
            if self.timeRemaining<=0:
                self.currentTask=None
    def busy(self):
        if self.currentTask!=None:
            return True
        else:
            return False
    def starNext(self,newtask):
        self.currentTask=newtask
        self.timeRemaining=newtask.getPages()/self.speed*60

class Task:
    def __init__(self,time):
        self.timestamp=time
        self.pages=random.randrange(1,21)
    def getPages(self):
        return self.pages
    def getStamp(self):
        return self.timestamp
    def waitTime(self,currentTime):
        return currentTime-self.timestamp
        
def newPrint():
    num=random.randrange(0,180)
    if num==0:
        return True
    else:
        return False

def simulation(numSeconds,pagesPerMinute):
    printer=Printer(pagesPerMinute)
    printqueue=Queue()
    waitingtimes=[]
    for currentSecond in range(numSeconds):
        if newPrint():
            task=Task(currentSecond)
            printqueue.enqueue(task)
        if not printer.busy() and not printqueue.isEmpyt():
            new=printqueue.dequeue()
            waitingtimes.append(new.waitTime(currentSecond))
            printer.starNext(new)
        printer.tick()
    if len(waitingtimes) > 0:
            averageWait = sum(waitingtimes)/len(waitingtimes)
    else:
            averageWait = 0 
    print(f'average wait{averageWait} secs {printqueue.size()} is remaining' )

for i in range(10):
    simulation(3600,5)
```

```python
# 构建双端队列
class Deque:
    def __init__(self):
        self.items=[]
    def addFront(self,item):
        self.items.append(item)
    def addRear(self,item):
        self.items.insert(0,item)
    def removeFront(self):
        return self.items.pop()
    def removeRear(self):
        return self.items.pop(0)
    def isEmpty(self):
        return self.items==[]
    def size(self):
        return len(self.items)

# 判断回文词

def judgehuiwen(n):
    deque=Deque()
    for i in n:
        # print(i)
        deque.addFront(i)
    # print(deque.items)
    if len(n)%2==0:
        while not deque.isEmpty():
            if deque.removeFront()!=deque.removeRear():
                return False
    else:
        while deque.size()!=1:
            if deque.removeFront()!=deque.removeRear():
                return False
    return True
print(judgehuiwen(input()))
            
        
        
```

```python
# 链表构造
class Node:
    def __init__(self,val):
        self.data=val
        self.next=None
    def getData(self):
        return self.data
    def getNext(self):
        return self.next
    def setData(self,newdata):
        self.data=newdata
    def setNext(self,newnext):
        self.next=newnext

class nodeList():
    def __init__(self):
        self.head=None

    def add(self,item):
        temp=Node(item)
        temp.setNext(self.head)
        self.head=None
    def size(self):
        current=0
        temp=self.head
        while not temp:
            current+=1
            temp=temp.getNext()
        return current
    def search(self,val):
        temp=self.head
        while not temp:
            if temp.getData()==val:
                return True
            else:
                temp=temp.getNext()
        return False
    def remove(self,val):
        temp=self.head
        previous=None
        found=False
        while not found:
            if temp.getData()==val:
                found=True
            else: 
                previous=temp
                temp=temp.getNext()
        if previous==None:
            self.head=temp.getNext()
        else:
            previous.setNext(temp.getNext())
            
```
