```python
from typing import List
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        m=len(matrix)
        n=len(matrix[0])
        total=m*n
        out=[]
        i=0
        j=0
        while len(out)!=total:
            while j<=n-1 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                j+=1
            # j,i
            j-=1
            i+=1
            while i<=m-1 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                i+=1
            i-=1
            j-=1
            # 
            while j>=0 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                j-=1
            j+=1
            i-=1
            # 
            while i>=0 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                i-=1
            i+=1
            j+=1
        return out
                
                
            
```

```python
a=Solution()
matrix=[[2,5],
        [8,4],
        [0,-1]]
a.spiralOrder(matrix)
```
