给定一个 n × n 的二维矩阵 matrix 表示一个图像。请你将图像顺时针旋转 90 度。

你必须在 原地 旋转图像，这意味着你需要直接修改输入的二维矩阵。请不要 使用另一个矩阵来旋转图像。


思路分析
==
感觉和那个螺旋矩阵的思路是一样的的，无非是多了一个移动  
然后奇偶数矩阵的判断要多一个步骤  
诶 不对，不是平移一格，是要找到对应的格子，得一次性把四个格子都找出来，一层一层变换

```python
from typing import List
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        # DIRS=(0,1),(1,0),(0,-1),(-1,0)
        m=len(matrix)
        n=len(matrix)//2
        step_max=m*n
        step=0
        # dirs=0
        while step<n:
            for i in range(step,m-step-1):
                # 右边那列的坐标
                right_row=i
                right_col=m-step-1
                # 底边的坐标
                bottom_row=m-step-1
                bottom_col=m-1-i 
                # 左边的坐标
                left_row=m-i-1
                left_col=step
                temp=matrix[step][i]
                matrix[step][i]=matrix[left_row][left_col]
                matrix[left_row][left_col]=matrix[bottom_row][bottom_col]
                matrix[bottom_row][bottom_col]=matrix[right_row][right_col]
                matrix[right_row][right_col]=temp
            step+=1
            
   
```

正确思路
==
题解是说一次转置加一次行翻转算一次顺时针旋转，不是哥们，这个不看题解谁能想得到啊
  
只能说牛逼

```python
from typing import List
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        for i in range(len(matrix)):
            # 这个地方倒是一个奇怪的小细节，范围是i
            # 其实这一整块的matrix[i][j]可以理解为对角线上方的元素
            for j in range(i):
                matrix[i][j]=matrix[j][i]
        for row in matrix:
            row=row.reverse()
            
```
