给你一个 m 行 n 列的矩阵 matrix ，请按照 顺时针螺旋顺序 ，返回矩阵中的所有元素。



思路分析
==
目前感觉最大的难点就是有可能不知道什么时候转向或者是什么时候停下  
那我觉得解决方案就是把读取过的元素全部删掉，这样碰到空值就转向

```python
from typing import List
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        m=len(matrix)
        n=len(matrix[0])
        total=m*n
        out=[]
        i=0
        j=0
        while len(out)!=total:
            while j<=n-1 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                j+=1
            # 先回退j,进位i
            j-=1
            i+=1
            while i<=m-1 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                i+=1
            i-=1
            j-=1
            # 从右往左走
            while j>=0 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                j-=1
            j+=1
            i-=1
            # 从下往上走
            while i>=0 and matrix[i][j] is not None:
                out.append(matrix[i][j])
                matrix[i][j]=None
                i-=1
            i+=1
            j+=1
        return out
                
                
            
```

```python
a=Solution()
matrix=[[2,5],
        [8,4],
        [0,-1]]
a.spiralOrder(matrix)
```
