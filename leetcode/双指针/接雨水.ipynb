给定 n 个非负整数表示每个宽度为 1 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。


思路分析
==
传说当中华为的保洁阿姨都需要手搓的接雨水，让我看看是怎么个事  
我有一个绝妙的想法，就是一层一层计算雨水，先算最底层，找到最左和最右两个坐标，计算中间的宽度，然后扣掉有各自的地方，然后计算第二层，一直算到最高层

办法是可行的 就是太慢，过不了时间限制

```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        ans=0
        n=len(height)
        h=max(height)
        l1=list(item for item in range(len(height)))
        for i in range(1,h+1):
            count=sum(1 for j in height if j>=i)
            # 如果该高度没有两个柱子，直接退出
            if count<2:
                break
            # 其实这个地方是可以优化的我猜，因为每次要遍历的列表都是在缩小
            for index in l1[:]:
                if height[index]<i:
                    l1.remove(index)
            # 留下来的是所有高度大于i的索引
            # 这一层的面积是
            ans+=(max(l1)-min(l1)-1)-(len(l1)-2)
    
        return ans
            
            
        
```

正确思路
==
用一个巧妙的双指针实现了前后缀分解，得到该柱子前后最高柱子的高度，这个前后最高柱子思路挺神奇的，感觉很多地方可以借鉴，之前算最大面积其实也用了这个思路  
所以其实这类问题的核心思路很多倒是找到前一个最高柱子和后一个最高柱子，这是一个很关键的思路

```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        n=len(height)
        pre_max=[0]*n
        pre_max[0]=height[0]
        suf_max=[0]*n
        suf_max[n-1]=height[n-1]
        square=0
        for i in range(1,n):
            pre_max[i]=max(pre_max[i-1],height[i])
        for i in range(n-2,-1,-1):
            suf_max[i]=max(suf_max[i+1],height[i])
        for i in range(n):
            square+=min(suf_max[i],pre_max[i])-height[i]
        return square
            
```

正确思路2-相向双指针
==
这个思路倒是神奇  
创造组有两个指针和左右两个最大墙
左边墙低于右边墙的时候，左边指针的水位线上限就被确定了
右边墙更低的时候，右边墙的上限也被定死了，这个思路牛的

```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        left,right=0,len(height)-1
        pre_max=suf_max=ans=0
        while left<right:
            pre_max=max(pre_max,height[left])
            suf_max=max(suf_max,height[right])
            if pre_max>suf_max:
                ans+=suf_max-height[right]
                right-=1
            else:
                ans+=pre_max-height[left]
                left+=1
        return ans
```

正确思路3-单调栈
==
这个思路感觉像我的初始思路，一层一层把空隙填满，栈这个数据结构还是能玩出太多花活了

```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        st=[]
        ans=0
        for i,h in enumerate(height):
            while st and height[st[-1]]<=h:
                bottom=height[st.pop()]
                if not st:
                    break
                left=st[-1]
                ans+=(min(height[left],h)-bottom)*(i-left-1)
            st.append(i)
        return ans
                
```
