给定一个长度为 n 的整数数组 height 。有 n 条垂线，第 i 条线的两个端点是 (i, 0) 和 (i, height[i]) 。

找出其中的两条线，使得它们与 x 轴共同构成的容器可以容纳最多的水。

返回容器可以储存的最大水量。



思路分析
==
这道题和那个最大矩形有点像，那个是用栈的做法，但这个让用双指针，让我想一想  有什么可以用不遍历所有可能性就找到最大面积的办法呢

我有一个绝妙的想法  
首先把指针放在列表头尾，算出一个值，然后指针往内移动，如果新的点的高度比原来还低，就直接跳过，因为底边变短了高度还没增加，不可能面积更大，但是我不太确定这种方法会不会把最大值给漏掉

```python
from typing import List
class Solution:
    def maxArea(self, height: List[int]) -> int:
        if len(height)<=1:
            return None
        right=len(height)-1
        left=0
        res=0
        while left<right:
            res=max(res,(right-left)*min(height[right],height[left]))
            print(f'res is {res},left is {left},right is {right}')
            if height[right]>=height[left]:
                temp=height[left]
                while temp>=height[left] and left<right:
                    left+=1
            else:
                temp=height[right]
                while temp>=height[right] and left<right:
                    right-=1
        return res
```

```python
a=Solution()
print(a.maxArea([1,8,6,2,5,4,8,3,7]))
```
