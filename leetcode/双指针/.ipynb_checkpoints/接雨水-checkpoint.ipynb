```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        ans=0
        n=len(height)
        h=max(height)
        l1=list(item for item in range(len(height)))
        for i in range(1,h+1):
            count=sum(1 for j in height if j>=i)
            # 
            if count<2:
                break
            # 
            for index in l1[:]:
                if height[index]<i:
                    l1.remove(index)
            # i
            # 
            ans+=(max(l1)-min(l1)-1)-(len(l1)-2)
    
        return ans
            
            
        
```

```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        n=len(height)
        pre_max=[0]*n
        pre_max[0]=height[0]
        suf_max=[0]*n
        suf_max[n-1]=height[n-1]
        square=0
        for i in range(1,n):
            pre_max[i]=max(pre_max[i-1],height[i])
        for i in range(n-2,-1,-1):
            suf_max[i]=max(suf_max[i+1],height[i])
        for i in range(n):
            square+=min(suf_max[i],pre_max[i])-height[i]
        return square
            
```

```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        left,right=0,len(height)-1
        pre_max=suf_max=ans=0
        while left<right:
            pre_max=max(pre_max,height[left])
            suf_max=max(suf_max,height[right])
            if pre_max>suf_max:
                ans+=suf_max-height[right]
                right-=1
            else:
                ans+=pre_max-height[left]
                left+=1
        return ans
```

```python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        st=[]
        ans=0
        for i,h in enumerate(height):
            while st and height[st[-1]]<=h:
                bottom=height[st.pop()]
                if not st:
                    break
                left=st[-1]
                ans+=(min(height[left],h)-bottom)*(i-left-1)
            st.append(i)
        return ans
                
```
