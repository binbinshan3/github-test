```python
from typing import List
class Solution:
    def maxArea(self, height: List[int]) -> int:
        if len(height)<=1:
            return None
        right=len(height)-1
        left=0
        res=0
        while left<right:
            res=max(res,(right-left)*min(height[right],height[left]))
            print(f'res is {res},left is {left},right is {right}')
            if height[right]>=height[left]:
                temp=height[left]
                while temp>=height[left] and left<right:
                    left+=1
            else:
                temp=height[right]
                while temp>=height[right] and left<right:
                    right-=1
        return res
```

```python
a=Solution()
print(a.maxArea([1,8,6,2,5,4,8,3,7]))
```
